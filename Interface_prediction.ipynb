```python
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import QSizePolicy,QLabel,QVBoxLayout
from PyQt5.QtGui import QDoubleValidator,QIntValidator
from PyQt5.QtGui import QFont
import sys
import numpy as np
import pickle
import tensorflow as tf
import pandas as pd

import os
import subprocess

class compil(object):
    def __init__(self):
       
        self.model = tf.keras.models.load_model("46loss76Acc.h5")
        self.arr=['','','','','','','','']
        

    
    def setupUi(self, compil):
        compil.setObjectName("compil")
        compil.setFixedSize(550, 800)
       
        #tous les elements de notre fenetre
        self.Heading = QtWidgets.QLabel(compil)
        self.Heading.setGeometry(QtCore.QRect(0, 10, 561, 41))
        self.Heading.setObjectName("Heading")

        self.layoutWidget = QtWidgets.QWidget(compil)
        self.layoutWidget.setGeometry(QtCore.QRect(50, 100, 350, 450))
        self.layoutWidget.setObjectName("layoutWidget")
        
        self.gridLayout = QtWidgets.QGridLayout(self.layoutWidget)
        self.gridLayout.setContentsMargins(0, 0, 0, 0)
        self.gridLayout.setObjectName("gridLayout")
        
        self.dp = QtWidgets.QLabel(self.layoutWidget)
        self.dp.setObjectName("dp")
        self.gridLayout.addWidget(self.dp, 2, 0, 1, 1)
        
        self.glucose_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.glucose_2.setObjectName("glucose_2")
        self.glucose_2.setValidator(QDoubleValidator(70.00, 180.00, 3))
        self.gridLayout.addWidget(self.glucose_2, 0, 1, 1, 1)
        self.glucose_2.setPlaceholderText("30-200 mg/dl")

        self.sk_thick = QtWidgets.QLabel(self.layoutWidget)
        self.sk_thick.setObjectName("sk_thick")
        self.gridLayout.addWidget(self.sk_thick, 7, 0, 1, 1)
        
        self.dp_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.dp_2.setObjectName("dp_2")
        self.gridLayout.addWidget(self.dp_2, 2, 1, 1, 1)
        self.dp_2.setPlaceholderText("diabetes pedigree index 0-3")
        self.dp_2.setValidator(QDoubleValidator(.00, 3.00, 3))
      
        self.bmi_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.bmi_2.setObjectName("bmi_2")
        self.gridLayout.addWidget(self.bmi_2, 5, 1, 1, 1)
        self.bmi_2.setPlaceholderText("0-70")
        self.bmi_2.setValidator(QDoubleValidator(10.00, 50.00, 2))
        
        self.preg_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.preg_2.setObjectName("preg_2")
        self.gridLayout.addWidget(self.preg_2, 1, 1, 1, 1)
        self.preg_2.setPlaceholderText("0 if male")
        self.preg_2.setValidator(QIntValidator())
        
        self.glucose = QtWidgets.QLabel(self.layoutWidget)
        self.glucose.setObjectName("glucose")
        self.gridLayout.addWidget(self.glucose, 0, 0, 1, 1)
        
        self.bp = QtWidgets.QLabel(self.layoutWidget)
        self.bp.setObjectName("bp")
        self.gridLayout.addWidget(self.bp, 6, 0, 1, 1)
        
        self.sk_thick_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.sk_thick_2.setObjectName("sk_thick_2")
        self.gridLayout.addWidget(self.sk_thick_2, 7, 1, 1, 1)
        self.sk_thick_2.setPlaceholderText("0-80mm")
        self.sk_thick_2.setValidator(QDoubleValidator(10.00, 50.00, 2))
        
        self.insu = QtWidgets.QLabel(self.layoutWidget)
        self.insu.setObjectName("insu")
        self.gridLayout.addWidget(self.insu, 3, 0, 1, 1)
        
        self.age_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.age_2.setObjectName("age_2")
        self.gridLayout.addWidget(self.age_2, 4, 1, 1, 1)
        self.age_2.setPlaceholderText("1-100 ")
        self.age_2.setValidator(QDoubleValidator(0, 110, 0))
        
        self.preg = QtWidgets.QLabel(self.layoutWidget)
        self.preg.setObjectName("preg")
        self.gridLayout.addWidget(self.preg, 1, 0, 1, 1)
        
        self.age = QtWidgets.QLabel(self.layoutWidget)
        self.age.setObjectName("age")
        self.gridLayout.addWidget(self.age, 4, 0, 1, 1)
        
        self.bp_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.bp_2.setObjectName("bp_2")
        self.gridLayout.addWidget(self.bp_2, 6, 1, 1, 1)
        self.bp_2.setPlaceholderText("44-220")
        self.bp_2.setValidator(QDoubleValidator(70.00, 150.00, 2))
        
        self.insu_2 = QtWidgets.QLineEdit(self.layoutWidget)
        self.insu_2.setObjectName("insu_2")
        self.gridLayout.addWidget(self.insu_2, 3, 1, 1, 1)
        self.insu_2.setPlaceholderText("0-400 ")
        self.insu_2.setValidator(QDoubleValidator(15.00, 300.00, 2))
        
        self.bmi = QtWidgets.QLabel(self.layoutWidget)
        self.bmi.setObjectName("BMI")
        self.gridLayout.addWidget(self.bmi, 5, 0, 1, 1)
        
        self.Prediction = QtWidgets.QPushButton(compil,clicked=self.execute_all)
        self.Prediction.setGeometry(QtCore.QRect(350, 600, 100, 30))
        self.Prediction.setObjectName("Prediction")
        
        self.Prediction1 = QtWidgets.QPushButton(compil,clicked=self.execute_all1)
        self.Prediction1.setGeometry(QtCore.QRect(250, 600, 100, 30))
        self.Prediction1.setObjectName("Prediction1")
        
        self.clear = QtWidgets.QPushButton(compil,clicked=self.clearall)
        self.clear.setGeometry(QtCore.QRect(100, 600, 100, 30))
        self.clear.setObjectName("clear")
        
        self.result = QtWidgets.QLabel(compil)
        self.result.setGeometry(QtCore.QRect(170, 660, 291, 61))
        self.result.setObjectName("result")
        
        self.result1 = QtWidgets.QLabel(compil)
        self.result1.setGeometry(QtCore.QRect(150, 690, 291, 61))
        self.result1.setObjectName("result1")
        
        self.translateUi(compil)
        QtCore.QMetaObject.connectSlotsByName(compil)
        
        compil.setTabOrder(self.glucose_2, self.preg_2)
        compil.setTabOrder(self.preg_2, self.dp_2)
        compil.setTabOrder(self.dp_2, self.insu_2)
        compil.setTabOrder(self.insu_2, self.age_2)
        compil.setTabOrder(self.age_2, self.bmi_2)
        compil.setTabOrder(self.bmi_2, self.bp_2)
        compil.setTabOrder(self.bp_2, self.sk_thick_2)
        compil.setTabOrder(self.sk_thick_2, self.Prediction)
        compil.setTabOrder(self.Prediction, self.Prediction1)
       
       #tous les labels
    def translateUi(self, compil):
        _translate = QtCore.QCoreApplication.translate
        compil.setWindowTitle(_translate("compil", "IA Project -2022- Asmae Mahjoubi"))
        
        self.Heading.setText(_translate("compil", "<html><head/><body><p align=\"center\"><span style=\" font-size:14pt; font-weight:600; color:#207b60;\">Prediction du diabète  </span></p><br></body></html>"))
        self.dp.setText(_translate("compil", "<html><head/><body><p align=\"right\">Diabets pedigree:</p></body></html>"))
        self.sk_thick.setText(_translate("compil", "<html><head/><body><p align=\"right\">Skin Thickness: </p></body></html>"))
        self.glucose.setText(_translate("compil", "<html><head/><body><p align=\"right\">Glucose:</p></body></html>"))
        self.bp.setText(_translate("compil", "<html><head/><body><p align=\"right\">Blood Pressure :</p></body></html>"))
        self.insu.setText(_translate("compil", "<html><head/><body><p align=\"right\">Insulin :</p></body></html>"))
        self.preg.setText(_translate("compil", "<html><head/><body><p align=\"right\">Pregnancies:</p></body></html>"))
        self.age.setText(_translate("compil", "<html><head/><body><p align=\"right\">Age:</p></body></html>"))
        self.bmi.setText(_translate("compil", "<html><head/><body><p align=\"right\">BMI:</p></body></html>"))
        self.Prediction.setText(_translate("compil", "Result Machine"))
        self.Prediction1.setText(_translate("compil", "Result Deep "))
        self.clear.setText(_translate("compil", "Clear"))
        self.result.setText(_translate("compil", "<html><head/><body><p align=\"center\"><span style=\" font-weight:600; font-style:italic; color:#a40000;\"> </span></p></body></html>"))
        self.result1.setText(_translate("compil", "<html><head/><body><p align=\"center\"><span style=\" font-weight:600; font-style:italic; color:#a40000;\"> </span></p></body></html>"))
        
            
    def execute_all(self):
        
       #Tous les params ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
        
        all_inputs    =[self.preg_2,self.glucose_2,self.bp_2,self.sk_thick_2,self.insu_2,self.bmi_2,self.dp_2,self.age_2]
        all_inputs,ee = self.champ_vide(all_inputs)
        if ee == 0:
            all_inputs    =list(map(float,all_inputs))
            re = self.restrict(all_inputs)
            if re == 0: 
                
                self.arr=all_inputs
                all_inputs    =np.array([np.array(all_inputs)])  
                output = self.check_all_inputs(all_inputs)
                print (output)
        
                if output==1:
                    
                    self.result.setText("* Vous etes diabétique ! *" )
                    self.result.setFont(QFont("Arial",11, weight=QFont.Bold))
                    
                else:
                    self.result.setText("* Vous n'avez rien! *" )
                    self.result.setFont(QFont("Arial",11, weight=QFont.Bold))
            else:
                self.showdialog("invalid values as input")
        else:
            self.showdialog("please first enter the values")
            
    def execute_all1(self):
        
       #Tous les params ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
        s       =self.model
        all_inputs    =[self.preg_2,self.glucose_2,self.bp_2,self.sk_thick_2,self.insu_2,self.bmi_2,self.dp_2,self.age_2]
        all_inputs,ee = self.champ_vide(all_inputs)
        if ee == 0:
            all_inputs    =list(map(float,all_inputs))
            re = self.restrict(all_inputs)
            if re == 0: 
                
                self.arr=all_inputs
                all_inputs    =np.array([np.array(all_inputs)])  
                output = self.check_all_inputs1(all_inputs)
                print (output)
        
                if output>0.5:
                    
                    self.result1.setText("* Vous etes diabétique à :  %.2f %% " % (float(100*output)))
                    self.result1.setFont(QFont("Arial",11, weight=QFont.Bold))
                    
                else:
                    self.result1.setText("* Vous n'avez rien à: %.2f %% " % (float(100-output)))
                    self.result1.setFont(QFont("Arial",11, weight=QFont.Bold))
            else:
                self.showdialog("invalid values as input")
        else:
            self.showdialog("please first enter the values")

            
    @staticmethod
    def check_all_inputs(data):
        
        df=pd.DataFrame(data=data,index=[0])
        p = pickle.load(open('Random.pickle', 'rb'))
       
        op = p.predict(df)
      
        return op[0]
    
    @staticmethod        
    def check_all_inputs1(data):
        
        df=pd.DataFrame(data=data,index=[0])
        model = tf.keras.models.load_model('46loss76Acc.h5')
       
        op = model.predict(df)
        print (op)
    
        return op[0]
        
            
    def restrict(self,all_inputs):
        comp_err = 0
        if all_inputs[0]>25:
            comp_err+=1
            self.preg_2.clear()
        
        if all_inputs[1]>200.00 or all_inputs[1]<44.00:
            comp_err+=1
            self.glucose_2.clear()
        
        if all_inputs[2]>200.00 or all_inputs[2]<24.00:
            comp_err+=1
            self.bp_2.clear()
        
        if all_inputs[3]>100.00 or all_inputs[3]<0.00:
            comp_err+=1
            self.sk_thick_2.clear()
        
        if all_inputs[4]>900.00 or all_inputs[4]<0.00:
            comp_err+=1
            self.insu_2.clear()
        
        if all_inputs[5]>70.00 or all_inputs[5]<0.00:
            comp_err+=1
            self.bmi_2.clear()
        
        if all_inputs[6]>3.00 or all_inputs[6]<0.00:
            comp_err+=1
            self.dp_2.clear()
        
        if all_inputs[7]>100.00 or all_inputs[7]<0.00 :
            comp_err+=1
            self.age_2.clear()
        return comp_err

    def champ_vide(self,all_inputs):
        l=[]
        empty=0
        for x in all_inputs:
            l.append(x.text())
        if '' in l :
            empty=1
        for i in range(len(l)):
            if l[i]=='.':
                all_inputs[i].clear()
                empty=1
        return l,empty

    def clearall(self):
        self.age_2.clear()
        self.bmi_2.clear()
        self.insu_2.clear()
        self.bp_2.clear()
        self.glucose_2.clear()
        self.dp_2.clear()
        self.sk_thick_2.clear()
        self.preg_2.clear()
        self.result.clear()
        self.result1.clear()
        self.arr=['','','','','','','','']

    def showdialog(self,s):
        msg = QtWidgets.QMessageBox()
        msg.setIcon(QtWidgets.QMessageBox.Information)
        msg.setText("Message Alert!")
        msg.setInformativeText("Additional information")
        msg.setWindowTitle("Attention!")
        msg.setDetailedText(s)
        msg.setStandardButtons(QtWidgets.QMessageBox.Ok | QtWidgets.QMessageBox.Cancel)
        msg.exec_()
    

#main call to  program
if __name__== '__main__':
    app = QtWidgets.QApplication(sys.argv)
    Dialog = QtWidgets.QDialog()
    ui = compil()
    ui.setupUi(Dialog)
    Dialog.show()
    sys.exit(app.exec_())
```
